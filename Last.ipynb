```python
def num5(n):
    return '0' * (5 - len(str(n))) + str(n) # 5 -> 00005, A01234: 1234 -> 01234
```

```python
def average(arr):
    return sum(arr)/len(arr)

def var(arr):
    m = average(arr)
    summ = 0
    for i in arr:
        summ += (i - m)**2
    return summ/(len(arr) - 1)
    
def median(arr):
    brr = list()
    for i in arr:
        brr.append(i)
    brr.sort()
    l = len(arr)
    if l % 2 == 0:
        return (arr[l//2] + arr[l//2 - 1])/2
    else:
        return arr[l//2]
```

```python
def independence(data):
    m = median(data) 
    Rnk = list()
    #print(len(data))
    for i in range(len(data)):
        if data[i] > data[i - 1]:
            Rnk.append(1)
        elif data[i] < data[i - 1]:
            Rnk.append(-1)
        else:
            Rnk.append(0)
    n = len(data)
    nu = 1
    maximum = 1
    count = 1
    for i in range(1, len(Rnk)):
        if Rnk[i] * Rnk[i - 1] == -1 or np.abs(Rnk[i] + Rnk[i - 1]) == 1:
            nu += 1
        if Rnk[i] == Rnk[i - 1]:
            count += 1
        else:
            if count > maximum:
                maximum = count
            count = 1
    #print(nu, maximum)
    if nu > 0.5 * (n + 1 - 2 * np.sqrt(n - 1)) and maximum < 3.3 * log(n + 1, 10):
        return True
    else:
        return False

def abbe(data):
    gamma = 0
    n = len(data)
    for i in range(1, len(data)):
        gamma += (data[i] - data[i - 1])**2
    gamma /= 2 * (len(data) - 1) * var(data)
    return gamma/(np.sqrt((n - 2)/(n**2 - 1)))
```

```python
def rank(arr):
    brr = [0] * len(arr)
    maxs = set()
    #topN = list()
    ler = 0
    while len(maxs) != len(arr):
        maximum = -10000
        #print(maxs)
        I = 0
        for i in set(range(len(arr))) - maxs:
            if arr[i] > maximum:
                maximum = arr[i]
                I = i
        #ler += 1
        #print(I)
        #if ler > 0:
         #   topN.append(I)
        brr[I] = len(maxs)
        maxs.update({I})
    return brr
    
```

```python
def series(data):
    #x = average(data)
    #data = list(map(lambda a: a - x, data))
    ser = 1
    n1 = 0
    n2 = 0
    for i in range(1, len(data)):
        if data[i] * data[i - 1] <= 0:
            ser += 1
    for i in data:
        if i >= 0:
            n1 += 1
        else:
            n2 += 1
    #print(ser)
    E = 2 * n1 * n2/(n1 + n2) + 1
    Var = (2 * n1 * n2/(n1 + n2)**2) * (2 * n1 * n2 - n1 - n2)/(n1 + n2 - 1)
    #print(E, Var)
    return (ser - E)/np.sqrt(Var)

def valdvolf(data):
    arr = rank(data)
    n = len(data)
    summ = 0
    for i in range(n - 1):
        summ += (arr[i] - (n + 1)/2) * (arr[i - 1] - (n + 1)/2)
    Var = n**2 * (n + 1) * (n - 3) * (5 * n + 6)/720
    return summ/np.sqrt(Var) + 1.1216 * n**(-0.523)

def autocor(data):
    n = len(data)
    sum1 = 0
    for i in range(1, len(data)):
        sum1 += data[i - 1] * data[i]
    sum1 *= n
    sum2 = sum(data)**2
    numerator = sum1 - sum2 + data[0] * data[-1]
    sum3 = reduce(lambda x, y: x**2 + y**2, data) * n
    denominator = sum3 - sum2
    R = numerator/denominator
    E = -1/(n - 1)
    Var = n * (n - 3)/((n + 1) * (n - 1)**2)
    return (R - E)/Var

def otvet(data):
    X = average(data)
    V = var(data)
    count = 0
    for i in data:
        if i > X + 1.6 * np.sqrt(V) or i < X - 1.6 * np.sqrt(V):
            count += 1
    return count
```
